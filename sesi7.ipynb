### Histogram
Histogram digunakan untuk mempresentasikan distribusi frekuensi dari dataset numerik. Cara kerjanya dalah mempartisi sumbu-x menjadi beberapa bin, menetapkan setiap titik data dalam kumpulan data ke bin, dan kemudian menghitung jumlah titik data yang telah ditetapkan ke setiap bin. Jadi, sumbu-x adalah frekuensi atau banyak titik pada data setiap bin

#### <b>Studi Kasus :</b> Berapa banyak frekuensi distribusi dari number (populasi) dari imigran baru dari negara-negara ke Canada di tahun 2013?

```python
df_can['2013'].head()
```

```python
count, bin_edges = np.histogram(df_can['2013'])
print(count)
print(bin_edges)
```

```python
df_can['2013'].plot(kind='hist', figsize=(8, 5))

plt.title('Histogram of Immigration from 195 Countries in 2013')
plt.ylabel('Number of Countries')
plt.xlabel('Number of Immigrants')
plt.show()
```

```python
# Memperbaiki ukuran bin dengan xticks

count, bin_edges = np.histogram(df_can['2013'])
df_can['2013'].plot(kind='hist', figsize=(8, 5), xticks=bin_edges)

plt.title('Histogram of Immigration from 195 Countries in 2013')
plt.ylabel('Number of Countries')
plt.xlabel('Number of Immigrants')
plt.show()
```

### Pie Charts
Pie chart adalah grafik circular yang digunakan untuk menampilkan proporsi numerik dengan membagi lingkaran (pie) menjadi irisan proporsional.

```python
# Step 1. Gather data : meringkas data imigrasi berdasarkan Continent
df_continents = df_can.groupby('Continent').sum()
print(type(df_can.groupby('Continent'))) 
df_continents.head()
```

```python
# Step 2. Plotting data
df_continents['Total'].plot(kind='pie', figsize=(5, 6),
                            autopct='%1.1f%%', startangle=90, shadow=True)
plt.title('Immigration to Canada by Continent [1980 - 2013]')
plt.axis('equal')
plt.show()
```

```python
# Step 3. Merapikan tampilan visual
colors_list = ['gold', 'yellowgreen', 'lightcoral', 
               'lightskyblue','lightgreen', 'pink']
explode_list = [0.1, 0, 0, 0, 0.1, 0.1] # rasio offset untuk setiap continent

df_continents['Total'].plot(kind='pie', figsize=(13, 8),
                            autopct='%1.1f%%', startangle=90, 
                            shadow=True, labels=None, pctdistance=1.12,
                            colors=colors_list, explode=explode_list)
plt.title('Immigration to Canada by Continent [1980 - 2013]')
plt.axis('equal')
plt.legend(labels=df_continents.index, loc='upper left')
plt.show()
```

### Box Plots
Box plot adalah cara untuk secara statistik mempresentasikan distribusi data melalui 5 dimensi utama, yaitu :
1. Minimum : Nilai terkecil dalam dataset
2. First Quartile : Nilai tengah antara minimum dan median
3. Second Quartile (Median) : Nilai tengah dari pengurutan dataset
4. Third Quartile : Nilai tengah diantara median dan maximum
5. Maximum : Nilai tertinggi dalam dataset

```python
# Step 1 : Menyiapkan dataframe
df_japan = df_can.loc[['Japan'], years].transpose()
df_japan.head()
```

```python
# Step 2 : Plotting
df_japan.plot(kind='box', figsize=(8, 6))
plt.title('Box plot of Japanese Immigrants from 1980 - 2013')
plt.ylabel('Number of Immigrants')
plt.show()
```

<b>Insight :</b> Berdasarkan plot, terdapat beberapa observasi penting, yaitu :
1. Nilai minimum imigran sekitar 200, maksimumnya sekitar 1300, dan mediannya sekitar 900 imigran
2. 25% dari 1980 - 2013 mempunyai sekitar 500 imigran (quartile 1)
3. 75% dari 1980 - 2013 mempunyai sekitar 1100 imigran (quartile 3)

### Bubble Plots
Bubble plot adalah variasi dari scatter plot yang menampilkan 3 dimensi data (x, y, z). Titik data diganti dengan bubbles dan ukuran bubbles ditentukan oleh variabel ketiga (z) yan juga dikenal dengan bobot

```python
df_can_t = df_can[years].transpose()
df_can_t.index = map(int, df_can_t.index)
df_can_t.index.name = 'Year'
df_can_t.reset_index(inplace=True)
df_can_t
```

<b> Normalisasi Data</b><br>
Feature scaling dapat digunakan untuk membuat normasilasi data untuk membawa semua nilai ke dalam kisaran [0, 1]. Rumus umumnya adalah X' = (X - Xmin) / (Xmax - Xmin)
<br>
dimana : X adalah nilai original dan X' adalah nilai yang dinormalisasi

```python
# Normalisasi data Brazil
norm_brazil = (df_can_t['Brazil'] - df_can_t['Brazil'].min()) / (df_can_t['Brazil'].max() - df_can_t['Brazil'].min())
                
# Normalisasi data Argentina
norm_argentina = (df_can_t['Argentina'] - df_can_t['Argentina'].min()) / (df_can_t['Argentina'].max() - df_can_t['Argentina'].min())
```

```python
norm_brazil
```

```python
norm_argentina
```

```python
# Brazil 
ax0 = df_can_t.plot(kind='scatter', x='Year', y='Brazil',
                    figsize=(14, 8), alpha=0.5, color='green',
                    s=norm_brazil * 2000 + 10, xlim=(1980, 2013))

# Argentina
ax1 = df_can_t.plot(kind='scatter', x='Year', y='Argentina',
                    figsize=(14, 8), alpha=0.5, color='blue',
                    s=norm_argentina * 2000 + 10, ax = ax0)

ax0.set_ylabel('Number of Immigrants')
ax0.set_title('Immigration from Brazil and Argentina from 1980 - 2013')
ax0.legend(['Brazil', 'Argentina'], loc='upper left', fontsize='x-large')
```
